```python
import pandas as pd

list_of_clusters = ["cluster_0", "cluster_1", "cluster_2", "cluster_3"]

allcounties = pd.read_csv('notebooks/clusterResults_0.csv')

cols_to_norm = ['Donald Trump','John Kasich', 'Ted Cruz']
allcounties["sum"] = allcounties[cols_to_norm].sum(axis=1)
allcounties[cols_to_norm] = allcounties[cols_to_norm].div(allcounties["sum"], axis="index")
# allcounties["sum"] = allcounties[cols_to_norm].sum(axis=1)
allcounties["winner"] = allcounties[cols_to_norm].idxmax(axis=1)

allcounties

for i in list_of_clusters[1:]:
    cluster_temp = i
    cluster_temp = pd.read_csv('notebooks/clusterResults_'+ i[-1] +'.csv')

    cols_to_norm = ['Donald Trump','John Kasich', 'Ted Cruz']
    cluster_temp["sum"] = cluster_temp[cols_to_norm].sum(axis=1)
    cluster_temp[cols_to_norm] = cluster_temp[cols_to_norm].div(cluster_temp["sum"], axis="index")
#     cluster_temp["sum"] = cluster_temp[cols_to_norm].sum(axis=1)
    cluster_temp["winner"] = cluster_temp[cols_to_norm].idxmax(axis=1)
    
    allcounties = pd.concat([allcounties, cluster_temp])

allcounties.to_csv('county_predictions.csv')
```

```python
county_pop_2012 = pd.read_csv("notebooks/county_facts_with_2012_winner.csv")
county_pop_2012 = county_pop_2012[["fips", "Paul", "Gingrich", "Romney", "Santorum", "state_abbreviation"]]
county_pop_2012["population"] = county_pop_2012[["Paul", "Gingrich", "Romney", "Santorum"]].sum(axis=1)
county_pop_2012 = county_pop_2012[["fips", "population", "state_abbreviation"]]
county_pop_percentage = pd.merge(allcounties, county_pop_2012, how="left", on="fips")

county_pop_percentage[cols_to_norm] = county_pop_percentage[cols_to_norm].apply(lambda x: x*county_pop_percentage["population"], axis=0)
county_pop_percentage = county_pop_percentage.groupby("state_abbreviation").sum()
# county_pop_percentage = county_pop_percentage.groupby("state_abbreviation").sum()[cols_to_norm].idxmax(axis=1)
county_pop_percentage[cols_to_norm] = county_pop_percentage[cols_to_norm].apply(lambda x: x/county_pop_percentage["population"], axis=0)
county_pop_percentage
```
